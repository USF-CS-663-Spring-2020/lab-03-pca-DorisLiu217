# 1. WDBC dataset

```python
import pandas as pd
df_wdbc = pd.read_csv('wdbc.data', header = None)
df_wdbc.head()
```

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']
df_wdbc.columns = column_labels
```

```python
df_wdbc['diagnosis'].replace('B', 'Benign', inplace=True)
df_wdbc['diagnosis'].replace('M', 'Malignant', inplace=True)
df_wdbc
```

```python
feature_df_wdbc = df_wdbc.iloc[:, 2:]
label_df_wdbc = df_wdbc.iloc[:, 1]
```

```python
feature_df_wdbc
```

```python
label_df_wdbc
```

```python
from sklearn.preprocessing import StandardScaler

x_wdbc = feature_df_wdbc.values
x_wdbc = StandardScaler().fit_transform(x_wdbc)
x_wdbc
```

```python
import numpy as np
np.mean(x_wdbc), np.std(x_wdbc)
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components = 2)
principalComponents_breast = pca_breast.fit_transform(x_wdbc)
principal_breast_df = pd.DataFrame(data = principalComponents_breast, columns = ['principal component 1', 'principal component 2'])
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
principal_breast_df
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = label_df_wdbc == target
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

# 2. Gisette  train dataset

```python
df_train = pd.read_csv('gisette_train.data', header = None, delim_whitespace=True)
df_train.head()
```

```python
label_train = pd.read_csv('gisette_train.labels', header = None).iloc[:, 0]
label_train.head()
```

```python
x_train = df_train.values
x_train = StandardScaler().fit_transform(x_train)
x_train
```

```python
np.mean(x_train), np.std(x_train)
```

```python
pca_train = PCA(n_components = 2)
principalComponents_train = pca_train.fit_transform(x_train)
principal_train_df = pd.DataFrame(data = principalComponents_train, columns = ['principal component 1', 'principal component 2'])
```

```python
print('Explained variation per principal component: {}'.format(pca_train.explained_variance_ratio_))
```

```python
principal_train_df
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.title("Principal Component Analysis of Gisette Train Dataset",fontsize=20)
targets_train = [1, -1]
colors = ['b', 'r']
for target, color in zip(targets_train,colors):
    indicesToKeep = label_train == target
    plt.scatter(principal_train_df.loc[indicesToKeep, 'principal component 1']
               , principal_train_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets_train,prop={'size': 15})
```
